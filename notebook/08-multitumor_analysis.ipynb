```python
# This script is for analyzing multitumor run
# Author Yiyun

import pandas as pd
from os.path import join, isdir, exists
from os import listdir, remove
import numpy as np
import matplotlib.pyplot as plt
```

```python
dir_sout = '../mutsig_out/multitumor/out08262020/syn'
dir_nsout = '../mutsig_out/multitumor/out08262020/nsyn'
dir_so = '../mutsig_out/syn/out08262020/Ovary'
dir_st = '../mutsig_out/syn/out08262020/Thy'
dir_no = '../mutsig_out/nsyn/out08262020/Ovary'
dir_nt = '../mutsig_out/nsyn/out08262020/Thy'

def getdf_mm(dirs):
    for f in listdir(dirs):
        f_path = join(dirs,f)
        if isdir(f_path):
            for fs in listdir(f_path):
                if fs.endswith('sig_genes.txt'):
                    fsig_path = join(f_path,fs)
                    df = pd.read_csv(fsig_path,index_col=0 ,sep = '\t')
    return df

def getdf(dirs):
    for f in listdir(dirs):
        if f.endswith('sig_genes.txt'):
            fsig_path = join(dirs,f)
            df = pd.read_csv(fsig_path, index_col = 0, sep = '\t')
    return df
```

```python
dfot_syn = getdf_mm(dir_sout)
dfot_nsyn = getdf_mm(dir_nsout)
dfo_syn = getdf(dir_so)
dft_syn = getdf(dir_st)
dfo_nsyn = getdf(dir_no)
dft_nsyn = getdf(dir_nt)
```

```python
df_sp = pd.DataFrame({'Ovary_only':dfo_syn.iloc[1:30,12], 'Thyroid_only':dft_syn.iloc[1:30,12], 'Ovary_Thyroid': dfot_syn.iloc[1:30,12]})
df_sp.reset_index(level=0, inplace=True)
```

```python
df_np = pd.DataFrame({'Ovary_only':dfo_nsyn.iloc[1:30,12], 'Thyroid_only':dft_nsyn.iloc[1:30,12], 'Ovary_Thyroid': dfot_nsyn.iloc[1:30,12]})
df_np.reset_index(level=0, inplace=True)
```

```python
ax = plt.gca()

df_sp.plot(kind='line',x='index',y='Ovary_only',ax=ax, figsize=(20,10), style='.', ms =15)
df_sp.plot(kind='line',x='index',y='Thyroid_only', color='red',style='.', ax=ax, ms =15)
df_sp.plot(kind='line',x='index',y='Ovary_Thyroid', color='blue', ax=ax,style='.', ms =15)

ax.set_xticks(range(len(df_sp)));
ax.set_xticklabels([item for item in df_sp['index'].tolist()], rotation=90);

xticks = ax.get_xticks()
for xtick in xticks:
        ax.vlines(xtick,0,0.025,linewidth=0.5)
        
ax.legend(scatterpoints=1,loc='best',ncol=1,markerscale=0.5,fontsize=12)
ax.set_title('Synonymous top 30 sig genes p value', 
             fontweight ="bold") 

plt.show()
```

```python
ax = plt.gca()

df_np.plot(kind='line',x='index',y='Ovary_only',ax=ax, figsize=(20,10), style='.', ms =15)
df_np.plot(kind='line',x='index',y='Thyroid_only', color='red',style='.', ax=ax, ms =15)
df_np.plot(kind='line',x='index',y='Ovary_Thyroid', color='blue', ax=ax,style='.', ms =15)

ax.set_xticks(range(len(df_np)));
ax.set_xticklabels([item for item in df_np['index'].tolist()], rotation=90);

xticks = ax.get_xticks()
for xtick in xticks:
        ax.vlines(xtick,0,0.015,linewidth=0.5)
        
ax.legend(scatterpoints=1,loc='best',ncol=1,markerscale=0.5,fontsize=12)
ax.set_title('Non-synonymous top 30 sig genes p value', 
             fontweight ="bold") 

plt.show()
```
