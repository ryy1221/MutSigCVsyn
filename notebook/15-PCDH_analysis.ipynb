```python
def get_overlapping_cons(muts):
    ext = '/vep/human/region/'+muts+'/?'
    r = requests.get(server+ext,headers={"Content-Type" : "application/json"})

    if not r.ok:
      r.raise_for_status()
      sys.exit()

    decoded = r.json()
    
    return decoded[0]['transcript_consequences']
```
